```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
finviz_url = 'https://finviz.com/quote.ashx?t='
#tickers = ['META','AMZN','AAPL','ABNB','ADBE','ADI','ADP','ADSK','AEP','ALGN','AMAT','AMD','AMGN','BKNG','CDNS','CEG','CHTR','CMCSA','COST','CPRT','CRWD','CSCO','CSX','CTAS','CTSH','DDOG','DLTR'
#,'DOCU','DXCM','EA','EBAY','EXC','FAST','FISV','FTNT','GILD','GOOG','GOOGL','HON','IDXX','ILMN','INTC','INTU','ISRG','JD','KDP','KHC'
#,'KLAC','LCID','LRCX','LULU','MAR','MCHP','MDLZ','MELI','MNST','MRNA','MRVL','MSFT','MTCH','MU','NFLX','NTES','NVDA','NXPI','ODFL','OKTA','ORLY','PANW','PAYX','PCAR','PDD','PEP','PYPL','QCOM'
#,'REGN','ROST','SBUX','SGEN','SIRI','SNPS','SPLK','SWKS','TEAM','TMUS','TSLA','TXN','VRSK','VRSN','VRTX','WBA','WDAY','XEL','ZM','ZS']

tickers = ['AMZN']
```

```python
news_tables = {}
for ticker in tickers:
    url = finviz_url + ticker

    req = Request(url=url, headers={'user-agent': 'my-app'})
    response = urlopen(req)

    html = BeautifulSoup(response, features='html.parser')
    news_table = html.find(id='news-table')
    news_tables[ticker] = news_table

parsed_data = []
title_data = []
```

```python
for ticker, news_table in news_tables.items():

    for row in news_table.findAll('tr'):

        title = row.a.text
        date_data = row.td.text.split(' ')

        if len(date_data) == 1:
            time = date_data[0]
        else:
            date = date_data[0]
            time = date_data[1]
            

        parsed_data.append([ticker, date, time, title])
        title_data.append([title])

df = pd.DataFrame(parsed_data, columns=['ticker', 'date', 'time', 'title'])
```

```python
def cleanTwt(twt):
  twt = re.sub('#[A-Za-z0-9]+','',str(twt))
  twt = re.sub('@[A-Za-z0-9]+','',str(twt))
  twt = re.sub('RT[\s]+','',str(twt))
  twt = re.sub('\\n','',str(twt))
  twt = re.sub('https?:\/\/\S+','',str(twt))
  return twt
```

```python
df['clean_text']= df['title'].apply(cleanTwt)
df.head(5)
```

```python
df["date"] = df['date'].astype(str)+' '+df["time"]
```

```python
df = df.drop(['time'], axis = 1)
```
